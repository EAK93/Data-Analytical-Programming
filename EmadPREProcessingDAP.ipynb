```python
#Importing the Libraries necessary for Data Pre Processing
import pandas as pd;
import numpy as np;
```

```python
#Since the document is in the XLS format, I have imported the file using Pandas and created a data frame
df=pd.read_excel("/home/emad/Documents/DAP/Prelim Semiannual Report to FIOU - Dec 14 2015/Table 4/Table_4_January_to_June_2015_Offenses_Reported_to_Law_Enforcement_by_State_by_City_100,000_and_Over_in_Population.xls")
```

```python
#Looking the data, its highly unformatted with unwated information. These unwanted information will be 
#removed and data will be structured step by step. 
df.head(10)
```

```python
#The first thing is to rename the columns properly of the data set. I have used the Dataframe rename 
#function to change the column names 
df.rename(columns={'Table 4': 'State', 'Unnamed: 1': 'City','Unnamed: 2':'Year', 'Unnamed: 3':'Population',
                   'Unnamed: 4':'Violent\Crime','Unnamed: 5':'Murder','Unnamed: 6':'Rape Revised',
                   'Unnamed: 7':'Rape Legacy','Unnamed: 8':'Robbery','Unnamed: 9':'Aggravated\Assault',
                   'Unnamed: 10':'Property\Crime','Unnamed: 11':'Burglary','Unnamed: 12':'Larceny\Theft',
                   'Unnamed: 13':'Motor\Vehicle\Theft','Unnamed: 14':'Arson'}, inplace=True)
```

```python
#The columns are now having proper names in the dataframe. The output is below. But it has some extra columns too.
df.head(2)
```

```python
#The first four rows in the data set are of no use and they are messing up the whole data set, 
#therefore, I will remove the first four rows using the drop function. It can be seen that I am not using the 
#column names but rather using the index to save some writing.
df.drop(df.index[:4], inplace=True)
```

```python
#Similarly the last rows after the the index 519 are also not needed as they are not part of the data set. 
#I have removed them too.
df.drop(df.index[520:], inplace=True)
```

```python
#The unstructured data also had some of the extra rows that are not needed, They are also droped using the same 
#drop function but this time, its using the names of the columns in the data frame.
df.drop(['Unnamed: 15','Unnamed: 16','Unnamed: 17'], axis = 1, inplace = True)
```

```python
#Now the data set has proper column names and no unwanted information. But there is one issue now, that is 
#the indexing of the rows. Because I deleted the extra rows, the index didnt change. Therefore, I have to reindex
#them.
df.head(2)
```

```python
#Using the Reindex function on the data frame, I have reset the index of the data frame but when this funtion is 
#used it renames the old index into a column and that stays there. Therefore that will be deleted too below.
df = df.reset_index()
```

```python
#The index column needs to be deleted.
df.head(5)
```

```python
#The column is deleted using the delete function where the dataframes column that needs to be deleted is mentioned.
del df['index']
df.head(5)
```

```python
#The structure of the data is formed and now I can dive into the actual content of the data. First I wanted to see 
#what are the data types of the columns. In order to manipulate the properly each column should have a proper data
#type. I have used the Data types functon to see the list of data types in the data frame. Every column is an object
# except the year column which is a float. It does not make sense to use dates as float.
df.dtypes
```

```python
#Here I have used the apply function to set the data type of the year column to integer from float. It can be seen 
#below that the data type is now is integer.
df['Year'] = df['Year'].apply(np.int)
df['Year'].dtypes
```

```python
#Moving on to the State and City column. There is grouping going in the columns. But these groupings are unstructured
#and in the data frame, its having missing values. For analysis each column should be filled in. Therefore, I have 
#used forward filling method. Forward filling fills the next missing value with the previous non-missing value.
#Below is the code that is used to forward fill the City and State column
df['City'] = df['City'].fillna(method='ffill')
df['State'] = df['State'].fillna(method='ffill')
```

```python
df.head()
```

```python
#Now the next column to tackle is the population column. Values for 2014 are given but for 2015 are not present
#I will again use forward fillin method to fill in the values but then each value in 2015 row will be incremented
#with the percentage increase in population of United States.
df['Population']=df['Population'].fillna(method='ffill')
df.loc[df.Year == 2015,'Population'] = df['Population']+(df['Population']*0.073);
```

```python
#But the values are coming in decimal points.
df.head()
```

```python
#Converting the floats to integers in population column
df['Population']=df['Population']
df['Population'] = df['Population'].apply(np.int)
```

```python
#Now it can be seen below that the values are all uniform in the population column as well as the whole data set,
#meaning integers.
df.head()
```

```python
#Finally the data frame has now been exported to a csv file and it will be used in Azure Machine Learning Studio
#for the treatment of the missing values.
file_name="/home/emad/Documents/DAP/toBeTreated.csv"
df.to_csv(file_name, sep=',', encoding='utf-8')
```
